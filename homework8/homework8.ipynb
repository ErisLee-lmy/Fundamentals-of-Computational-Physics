```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 设置基本常数
Lx = 1#单位m
Ly = 1.5#单位m
dpi = 100 # 每m取点数
h = 1/dpi #步长
tol =1e-6#容忍度
max_inter = 10000

# 构建网格
x = np.linspace(0,Lx,int(Lx*dpi))
y = np.linspace(0,Ly,int(Ly*dpi))
X,Y = np.meshgrid(x,y)
Z = np.zeros_like(X)
# 设置边界条件
Z[:,0] = 0  
Z[:,-1] = 0
Z[0,:] = 0 
Z[-1,:] =1 #单位V

f  = np.zeros_like(Z)

def RelaxtionMethod(f,Z):
    '''输入泊松方程右侧的函数f和边界条件已经设置好的网格Z'''
    n_x  = np.shape(Z)[1]
    n_y = np.shape(Z)[0]
    for inter in range(max_inter):
        
        Z_new = np.copy(Z)
        for i in range(1,n_y-1):
            for j in range(1,n_x-1):
                Z_new[i,j] = 0.25*(Z[i+1,j]+Z[i-1,j]+Z[i,j+1]+Z[i,j-1]-h**2*f[i,j])
                
        err = np.max(np.abs(Z_new-Z))
        
        Z = Z_new
        
        if inter%100 == 0:
            print(f"已迭代{inter}次，尚未收敛")
        
        if err < tol:
            print(f"在迭代{inter}后结果收敛")
            break
        if inter == max_inter-1:
            print(f'迭代{max_inter}后仍未收敛,此时最大误差为{err}')
    return Z

Z = RelaxtionMethod(f,Z)
```

```python
# 绘制三维曲面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('Solution of Poisson Equation using Relaxation Method')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel(r'$\phi$ axis')
plt.show()

plt.contourf(X, Y, Z, 100, cmap='viridis')
plt.colorbar(label='$\phi(x, y)$')
plt.title('Solution of Poisson Equation using Relaxation Method')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 设置参数
L = 1# 方形井的宽度
V0 = 1# 井深
x_max = 20.0# 计算区域的右边界
Nx = 200 #x取点数
t_max = 10#时间最大值
Nt=1000#时间取点数
hbar = 1.0# 普朗克常数
m = 1.0#粒子质量

h_x = 2*x_max/Nx
h_t = t_max/Nt

# 网格和初始条件
x = np.linspace(-x_max,x_max,Nx)
t = np.linspace(0,t_max,Nt)

#势能
V = np.zeros_like(x)
V[np.abs(x)<=0.5*L] = -V0

# 定义初始波包
ksi0 = -5.0  # 初始位置
k0 = 1
psi0 = np.sqrt(1/np.pi)*np.exp(1j*k0*x-0.5*(x-ksi0)**2/2)

def CrankNicolson(V,psi0):
    Psi = np.zeros((Nt,Nx),dtype=complex)
    Psi[0,:] = psi0
    alpha = h_t/h_x**2# 离散化Crank-Nicolson方程的系数
    
    B = np.zeros((Nx,Nx))
    for a in range(Nx):
            for b in range(Nx):
                if a - b ==0:
                    B[a,b] = -0.5*V[a]*h_t/hbar-0.5*alpha*hbar/m
                if np.abs(a-b) == 1:
                    B[a,b] = 0.25*alpha*hbar/m
    
    for j in range(1,Nt):
        if (j+1)%10 ==0:
            print(f"已计算{j+1}个时间节点")
        
        
        Psi[j] = np.linalg.inv(1j*np.identity(Nx)+B)@(1j*np.identity(Nx)-B)@Psi[j-1]
        
    return Psi
Psi_CrankNicolson = CrankNicolson(V,psi0)
```

```python
fig = plt.figure(figsize=(10, 8))
ax= plt.axes(projection='3d')
# creating the plot 
X,T=np.meshgrid(x,t)
plot_geeks = ax.plot_surface(T,X,np.abs(Psi_CrankNicolson)**2, cmap='rainbow', edgecolor='none') 
ax.set_title("Solution for time-dependent  Schrödinger equatation under Crank–Nicolson scheme") 
ax.set_xlabel('t/s') 
ax.set_ylabel('x/m') 
ax.set_zlabel(r'$|\Psi|^2$',rotation=90) 
plt.show()
```

```python
def Explicit(V,psi0):
    alpha = h_t/(h_x**2)
    
    
    Psi = np.zeros((Nt,Nx))
    
    Psi[0] = psi0
    A = np.zeros((Nx, Nx), dtype=complex)
    np.fill_diagonal(A, -1j*hbar**2*alpha/m - 1j*h_t*V)  # 主对角线的值
    np.fill_diagonal(A[1:], 1j*hbar**2*alpha/m)  # 上对角线的值
    np.fill_diagonal(A[:, 1:], 1j*hbar**2*alpha/m)  # 下对角线的值
    Psi[1]=psi0+A@psi0
    
    A = np.zeros((Nx,Nx),dtype=complex)
    for a in range(Nx):
        for b in range(Nx):
            if a-b == 0:
                A[a,b] = -2j*hbar**2*alpha/m - 2j*h_t*V[a]
            if np.abs(a-b)==1:
                A[a,b] = 1j*hbar**2*alpha/m
    for j in range(2,Nt):
        Psi[j] = Psi[j-2]+A@Psi[j-1]
        if (j+1)%10 == 0:
            print(f"已计算{j+1}个时间节点")
    return Psi
Psi_Explicit = Explicit(V,psi0)
```

```python
fig = plt.figure(figsize=(10, 8))
ax= plt.axes(projection='3d')
# creating the plot 
X,T=np.meshgrid(x,t)
plot_geeks = ax.plot_surface(T,X,np.abs(Psi_Explicit)**2, cmap='rainbow', edgecolor='none') 
ax.set_title("Solution for time-dependent  Schrödinger equatation under  Stable Explicit Scheme") 
ax.set_xlabel('t/s') 
ax.set_ylabel('x/m') 
ax.set_zlabel(r'$|\Psi|^2$',rotation=90) 
plt.show()
```
